### Exercise 2B
The last slide introduces **parameter** vs **statistic**.

Treat the same grade list in two different ways:
- as the **entire population** of last year's class
- as a **sample** from all future classes

Compare the standard deviation in both cases.

```python
pd.DataFrame({
    "Interpretation": ["Population parameter", "Sample statistic"],
    "Standard deviation": [exam_scores.std(ddof=0), exam_scores.std(ddof=1)]
}).round(3)
```

## 3. What is Statistics? Data -> Information
A short Chapter 1 style reminder:

- **Descriptive statistics** summarize the data you have.
- **Inferential statistics** use sample data to say something about a larger population.

Let's practice that distinction before moving to a real dataset.

### Quick concept check (no code needed first)

1. "The average exam grade in this class was 5.75."
2. "Students in future classes will probably have an average close to this."
3. "The most common grade was 7."
4. "If we change the teaching method, future grades may improve."

```python
# Suggested answers
concept_check = pd.DataFrame({
    "Statement": [
        "The average exam grade in this class was 5.75.",
        "Students in future classes will probably have an average close to this.",
        "The most common grade was 7.",
        "If we change the teaching method, future grades may improve."
    ],
    "Likely type": [
        "Descriptive",
        "Inferential",
        "Descriptive",
        "Inferential"
    ]
})
concept_check
```
