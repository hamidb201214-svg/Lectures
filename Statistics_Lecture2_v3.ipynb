```python
import pandas as pd
import numpy as np

# Load the real Credit Card Default dataset directly from the public ISLR repository
url = "https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/ISLR/Default.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
# Convert 'default' to numeric (1 for Yes, 0 for No) so we can calculate percentages
df['default_rate'] = np.where(df['default'] == 'Yes', 1, 0)
```

```python
pd.crosstab(
    index=df['student'],
    columns='default_rate_mean',
    values=df['default_rate'],
    aggfunc='mean'
) * 100
```

```python
pd.crosstab(df['student'], df['default'], normalize='index') * 100
```

```python
df['balance_bucket'] = pd.cut(df['balance'],
                              bins=[0, 1000, 2000, 3000],
                              labels=['Low Debt ($0-$1k)', 'Medium Debt ($1k-$2k)', 'High Debt ($2k+)'])
```

```python
pd.crosstab(
    df['balance_bucket'],
    df['student'],
    values=df['default_rate'],
    aggfunc='mean'
) * 100
```
